```python
import pandas as pd

df = pd.read_json("transformed.json")

df['id'] = df.index.values
```

```python
def gen_data(columns):
    for col in columns:
        output = df[['id',col]].explode(column=col)
        output.to_csv(f"{col}.csv",index=False)
    
```

```python
gen_data(['entities','benefits','tasks'])
```
